```python
from selenium import webdriver
import time

def take_screenshot_while_scrolling(driver, output_folder="screenshots/aa"):
    import os
    
    # 폴더가 없는 경우 생성
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    total_height = driver.execute_script("return document.body.scrollHeight")  # 페이지의 총 높이를 가져옵니다
    viewport_height = driver.execute_script("return window.innerHeight")  # 현재 화면의 높이를 가져옵니다
    
    scroll_position = 0
    screenshot_count = 0
    
    while scroll_position < total_height:
        # 스크린샷 저장
        driver.save_screenshot(os.path.join(output_folder, f"screenshot_{screenshot_count}.png"))
        screenshot_count += 1
        
        # 스크롤 이동
        driver.execute_script("window.scrollBy(0, arguments[0])", viewport_height)
        scroll_position += viewport_height   # 현재 스크롤 위치 업데이트
        time.sleep(1)  # 스크롤 후 페이지가 로드되도록 잠시 대기
    
    print(f"총 {screenshot_count}개의 스크린샷이 저장되었습니다.")

# Use case
driver = webdriver.Edge()
driver.get("https://www.google.com/search?sca_esv=ef6a114c258b96d2&q=%EA%B9%80%EC%88%98%ED%98%84&udm=2&fbs=ABzOT_CZsxZeNKUEEOfuRMhc2yCIN42EXxa9ZSNEwtiPEbQrp-oREuj69PlSffsqaZff35ttlTfDht-WBlJ2aWSHHA1tvIKPilsjms6sA7mve6g0aZ2GNZxLK-Q9kz6iolKHL9x7PNRajm0mT6Y6sMnb8D9UCW0wyv1rkE_ljDJiyXkNlQtZzZ44Mrr5UNwmY8fwH8pFi0UvjHSUsYMfk8FPuqp0TqhTgQ&sa=X&sqi=2&ved=2ahUKEwjn0PrY0IOMAxW3zzQHHQAtDxIQtKgLegQIGhAB&biw=1280&bih=593&dpr=1.5")  # 스크린샷 찍을 페이지로 이동
time.sleep(3)  # 페이지 로드 대기

# 호출
take_screenshot_while_scrolling(driver)

driver.quit()
```

```python
# 특정요소가 로드 될때까지 대기(WebDriverWait)
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC # 대기 조건
from selenium.webdriver.common.by import By

driver=webdriver.Chrome()
driver.get("https://www.naver.com")

# name의 값이 'q' 인 엘리먼트가 로드될 때까지 10초간 기다림
# and element 선택이 됨
search_box=WebDriverWait(driver,10).until(
    EC.presence_of_element_located(('name',"query"))
)
#driver.find_element("")
search_box.send_keys("selenium 대기 테스트")
search_box.send_keys(Keys.ENTER)
time.sleep(5)
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import urllib.request


query = '장미'
driver = webdriver.Chrome()
driver.get(f'https://www.google.com/imghp')
# 구글 이미지 검색창 이름이 : q
search_bar = driver.find_element(By.NAME,"q")
# 검색 키워드를 춘식이 설정
search_bar.send_keys(query)
# 전달.
search_bar.submit()
# 잠시 대기 시간 ms 1000ms = 1초
PAUSE_TIME = 2
# 현재 문서의 스크롤의 높이를 표시.
last_hegiht = driver.execute_script("return window.scrollY")
print(f"처음 last_hegiht : {last_hegiht}")
new_height = 0

interval = 2
prev_height = driver.execute_script("return document.body.scrollHeight")

while True:
    #화면 가장 아래로 스크롤 내리기
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(interval)

    curr_height = driver.execute_script("return document.body.scrollHeight")
    if curr_height == prev_height:
        break

    prev_height = curr_height

img_elements = driver.find_elements(
    By.TAG_NAME,"img")
# 다시 이미지만 추출해서 담음 임시 리스트
print(f"img_elements ================================ : {img_elements}")

save_path = r'D:\pgm\bigdata_col\data\craw_img\{query}'
import os
if not os.path.exists(save_path):
    os.mkdir(save_path)


save_path = r'D:\pgm\bigdata_col\data\craw_img\{query}'
imgs = []
# enumerate 함수 이용하면 해당 인덱스도 같이 추출 가능.
for idx, img in enumerate(img_elements):
    print(f"{query} - {idx+1}/{len(img_elements)} 진행 중...")

    try:
        img_src = img.get_attribute("src")
        img_alt = img.get_attribute("alt")

        if img_src and "http" in img_src:  # src가 비어 있거나 base64가 아닌 경우 저장
            imgs.append({"alt": img_alt, "src": img_src})
            urllib.request.urlretrieve(img_src, f"{save_path}/{query}_{idx}.png")
            print(f"✅ {idx}: 다운로드 완료 - {img_alt}")

    except Exception as e:
        print(f"❌ {idx}: 다운로드 실패 - {e}")
        pass
driver.close()


print('done')
```
