```python
from selenium import webdriver
import time

def take_screenshot_while_scrolling(driver, output_folder="screenshots/aa"):
    import os
    
    # í´ë”ê°€ ì—†ëŠ” ê²½ìš° ìƒì„±
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    total_height = driver.execute_script("return document.body.scrollHeight")  # í˜ì´ì§€ì˜ ì´ ë†’ì´ë¥¼ ê°€ì ¸ì˜µë‹ˆë‹¤
    viewport_height = driver.execute_script("return window.innerHeight")  # í˜„ì¬ í™”ë©´ì˜ ë†’ì´ë¥¼ ê°€ì ¸ì˜µë‹ˆë‹¤
    
    scroll_position = 0
    screenshot_count = 0
    
    while scroll_position < total_height:
        # ìŠ¤í¬ë¦°ìƒ· ì €ì¥
        driver.save_screenshot(os.path.join(output_folder, f"screenshot_{screenshot_count}.png"))
        screenshot_count += 1
        
        # ìŠ¤í¬ë¡¤ ì´ë™
        driver.execute_script("window.scrollBy(0, arguments[0])", viewport_height)
        scroll_position += viewport_height   # í˜„ì¬ ìŠ¤í¬ë¡¤ ìœ„ì¹˜ ì—…ë°ì´íŠ¸
        time.sleep(1)  # ìŠ¤í¬ë¡¤ í›„ í˜ì´ì§€ê°€ ë¡œë“œë˜ë„ë¡ ì ì‹œ ëŒ€ê¸°
    
    print(f"ì´ {screenshot_count}ê°œì˜ ìŠ¤í¬ë¦°ìƒ·ì´ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")

# Use case
driver = webdriver.Edge()
driver.get("https://www.google.com/search?sca_esv=ef6a114c258b96d2&q=%EA%B9%80%EC%88%98%ED%98%84&udm=2&fbs=ABzOT_CZsxZeNKUEEOfuRMhc2yCIN42EXxa9ZSNEwtiPEbQrp-oREuj69PlSffsqaZff35ttlTfDht-WBlJ2aWSHHA1tvIKPilsjms6sA7mve6g0aZ2GNZxLK-Q9kz6iolKHL9x7PNRajm0mT6Y6sMnb8D9UCW0wyv1rkE_ljDJiyXkNlQtZzZ44Mrr5UNwmY8fwH8pFi0UvjHSUsYMfk8FPuqp0TqhTgQ&sa=X&sqi=2&ved=2ahUKEwjn0PrY0IOMAxW3zzQHHQAtDxIQtKgLegQIGhAB&biw=1280&bih=593&dpr=1.5")  # ìŠ¤í¬ë¦°ìƒ· ì°ì„ í˜ì´ì§€ë¡œ ì´ë™
time.sleep(3)  # í˜ì´ì§€ ë¡œë“œ ëŒ€ê¸°

# í˜¸ì¶œ
take_screenshot_while_scrolling(driver)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import urllib.request


query = 'ì¥ë¯¸'
driver = webdriver.Chrome()
driver.get(f'https://www.google.com/imghp')
# êµ¬ê¸€ ì´ë¯¸ì§€ ê²€ìƒ‰ì°½ ì´ë¦„ì´ : q
search_bar = driver.find_element(By.NAME,"q")
# ê²€ìƒ‰ í‚¤ì›Œë“œë¥¼ ì¶˜ì‹ì´ ì„¤ì •
search_bar.send_keys(query)
# ì „ë‹¬.
search_bar.submit()
# ì ì‹œ ëŒ€ê¸° ì‹œê°„ ms 1000ms = 1ì´ˆ
PAUSE_TIME = 2
# í˜„ì¬ ë¬¸ì„œì˜ ìŠ¤í¬ë¡¤ì˜ ë†’ì´ë¥¼ í‘œì‹œ.
last_hegiht = driver.execute_script("return window.scrollY")
print(f"ì²˜ìŒ last_hegiht : {last_hegiht}")
new_height = 0

interval = 2
prev_height = driver.execute_script("return document.body.scrollHeight")

while True:
    #í™”ë©´ ê°€ì¥ ì•„ë˜ë¡œ ìŠ¤í¬ë¡¤ ë‚´ë¦¬ê¸°
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(interval)

    curr_height = driver.execute_script("return document.body.scrollHeight")
    if curr_height == prev_height:
        break

    prev_height = curr_height

img_elements = driver.find_elements(
    By.TAG_NAME,"img")
# ë‹¤ì‹œ ì´ë¯¸ì§€ë§Œ ì¶”ì¶œí•´ì„œ ë‹´ìŒ ì„ì‹œ ë¦¬ìŠ¤íŠ¸
print(f"img_elements ================================ : {img_elements}")

save_path = r'D:\pgm\bigdata_col\data\craw_img\{query}'
import os
if not os.path.exists(save_path):
    os.mkdir(save_path)


save_path = r'D:\pgm\bigdata_col\data\craw_img\{query}'
imgs = []
# enumerate í•¨ìˆ˜ ì´ìš©í•˜ë©´ í•´ë‹¹ ì¸ë±ìŠ¤ë„ ê°™ì´ ì¶”ì¶œ ê°€ëŠ¥.
for idx, img in enumerate(img_elements):
    print(f"{query} - {idx+1}/{len(img_elements)} ì§„í–‰ ì¤‘...")

    try:
        img_src = img.get_attribute("src")
        img_alt = img.get_attribute("alt")

        if img_src and "http" in img_src:  # srcê°€ ë¹„ì–´ ìˆê±°ë‚˜ base64ê°€ ì•„ë‹Œ ê²½ìš° ì €ì¥
            imgs.append({"alt": img_alt, "src": img_src})
            urllib.request.urlretrieve(img_src, f"{save_path}/{query}_{idx}.png")
            print(f"âœ… {idx}: ë‹¤ìš´ë¡œë“œ ì™„ë£Œ - {img_alt}")

    except Exception as e:
        print(f"âŒ {idx}: ë‹¤ìš´ë¡œë“œ ì‹¤íŒ¨ - {e}")
        pass
driver.close()


print('done')
```
