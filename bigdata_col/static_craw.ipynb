```python
url="http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
data={'stnId':'159'}
r=requests.get(url,params=data)
r.encoding='utf-8'
webPage=r.text
soup=bs(webPage,'html.parser')
print(soup.prettify())
```

```python
list1=[]
loc=soup.find_all("location")
#print(loc)
for l in loc:
    city=l.city.string
    datas=l.find_all("data")
    #print(datas)
    for d in datas:
        tmef=d.tmef.string
        wf=d.wf.string
        tmn=d.tmn.string
        tmx=d.tmx.string
        #print(city,tmef,wf,tmn,tmx)
        list1.append((city,tmef,wf,tmn,tmx))
print(list1)
```

```python
import pandas as pd
df=pd.DataFrame(list1,columns=['city','tmef','wf','tmn','tmx'])
df.to_csv("data/weather.csv",index=False,encoding="utf-8-sig")
```
